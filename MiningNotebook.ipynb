### 2. Bulding the tree using Gain Ratio(C5.0).
C5.0 is a common algorithm for building decision trees, and the gain ratio is used in the construction of these trees. Gain ratio is a measure that helps to select the best splitting attribute at each node of the decision tree.

When constructing a decision tree, the algorithm considers different attributes or features at each node and chooses the one that maximizes the gain ratio. Gain ratio can be calculated by taking into account the information gain and the intrinsic information of the attribute.

**Note**: 
- **Information Gain (IG):** A measure of the effectiveness of an attribute in classifying the data. It is calculated by comparing the entropy of the dataset before and after the split based on an attribute. Entropy is a measure of impurity or disorder in a set of data.
- **Intrinsic Information (II):** This is a measure of the amount of uncertainty in a set of data.



**Explanation of the code:**

we loaded the library of C5.0 and some other libraries for printing the results, built a model using the c5.0 method which uses the C5.0 to construct the model on the selected trainData and we specified 'price~.' which means it would predict the price variable based on all the rest of the attributes.

### 3. Building the Tree using Gini Index(CART).
In classification using decision trees, the Gini index is a measure of impurity in a node. It evaluates how often a randomly chosen element would be incorrectly classified. The goal is to minimize the Gini index by making splits that result in more homogenous subsets. A lower Gini index indicates a more pure node, contributing to trying to find an accurate decision tree for classification.

Nodes with a Gini index close to 0 are considered pure, meaning most instances in the node belong to the same class.

**Explanation of the code:**

- Firstly, we load the rpart package for the CART algorithm and the rpart.plot for visualizing the decision tree we are going to create.
- using rpart() method we identifiead a variable 'fit.tree' which will contain our constructed model by applying CART algorithm on our trainData, setting method="class" to indicate that this is a classification task, and 'cp = 0.008' to set the complexity parameter, which is what controls the level of tree complexity and would help avoid overfitting, ad then printng the fit.tree which prints some details about the decision tree model.
- using rpart.plot() to see how our decision tree looks with the amazing visualization from the rpart package.
- checking variable importance for each attribute in the dataset to check which feature has the most impact  on our predictions and order them accordingly.
- predicting the class labels using the trained model 'fit.tree' o the testData.
- and Lastly, creating a confusion matrix to compare the predictions with the actual class labels.


**A. CART on the 70-30 split:**

```r
# For decision tree model
install.packages("rpart")
library(rpart)
# For data visualization
library(rpart.plot)


fit.tree = rpart(price ~ ., data=trainData1, method = "class", cp=0.008)
fit.tree

# Visualizing the unpruned tree
rpart.plot(fit.tree)

# Checking the order of variable importance
fit.tree$variable.importance

pred.tree = predict(fit.tree, testData1, type = "class")

table(pred.tree,testData1$price)
```


The tree starts with the root node (1) and branches based on various conditions until terminal nodes marked with "*". Each node provides information such as the number of observations (n), the number of misclassifications (loss), the predicted value (yval), and the class probabilities (yprob).

- **False Positive (FP):** 19 instances were incorrectly predicted as "High" class.

**Accuracy:**
   - Accuracy = (TP + TN) / (TP + TN + FP + FN)
   - Accuracy = (48 + 51) / (51 + 19 + 16 + 48) = 99 / 134 ≈ 0.7388 or 73.88%

**Precision (Positive Predictive Value):**
   - Precision = TP / (TP + FP)
   - Precision = 48 / (48 + 19) ≈ 0.7164 or 71.64%

**Recall (Sensitivity or True Positive Rate):**
   - Recall = TP / (TP + FN)
   - Recall = 48 / (48 + 16) ≈ 0.75 or 75%

**Specificity (True Negative Rate):**
   - Specificity = TN / (TN + FP)
   - Specificity = 51 / (51 + 19) ≈ 0.7286 or 72.86%

**Misclassification Rate:**
   - Misclassification Rate = (FP + FN) / (TP + TN + FP + FN)
   - Misclassification Rate = (19 + 16) / (51 + 19 + 16 + 48) = 35 / 134 ≈ 0.2612 or 26.12%

```r
# Function to calculate entropy
calculate_entropy <- function(class_probs) {
  -sum(class_probs * log2(class_probs), na.rm = TRUE)
}

# Function to calculate information gain
calculate_information_gain <- function(data, attribute, class_label) {
  # Calculate the entropy of the entire dataset
  entropy_d <- calculate_entropy(table(data[[class_label]]) / nrow(data))

  # Calculate the weighted sum of entropies for each partition
  partitions <- split(data, data[[attribute]])
  entropy_after_split <- sum(
    sapply(partitions, function(partition) {
      (nrow(partition) / nrow(data)) * calculate_entropy(table(partition[[class_label]]) / nrow(partition))
    })
  )

  # Calculate information gain
  information_gain <- entropy_d - entropy_after_split

  return(information_gain)
}

# Function to construct a decision tree
construct_decision_tree <- function(data, class_label, attributes) {
  # Base case: if all instances have the same class, create a leaf node
  if (length(unique(data[[class_label]])) == 1) {
    return(data[[class_label]][1])
  }

  # Base case: if there are no attributes left, create a leaf node with the majority class
  if (length(attributes) == 0) {
    majority_class <- names(sort(table(data[[class_label]]), decreasing = TRUE))[1]
    return(majority_class)
  }

  # Select the attribute with the highest information gain
  information_gains <- sapply(attributes, function(attribute) {
    calculate_information_gain(data, attribute, class_label)
  })
  selected_attribute <- attributes[which.max(information_gains)]

  # Create a tree node with the selected attribute
  tree <- list()
  tree$attribute <- selected_attribute
  tree$children <- list()

  # Recursively construct subtrees for each value of the selected attribute
  attribute_values <- unique(data[[selected_attribute]])
  for (value in attribute_values) {
    subset_data <- data[data[[selected_attribute]] == value, ]
    subset_data <- subset_data[, !(names(subset_data) %in% selected_attribute)]  # Remove the used attribute
    subset_attributes <- setdiff(attributes, selected_attribute)
    
    # Recursively construct the subtree
    subtree <- construct_decision_tree(subset_data, class_label, subset_attributes)
    
    # Add the subtree to the current node
    tree$children[[as.character(value)]] <- subtree
  }

  return(tree)
}

# Example usage:
# Assuming trainData1 is your training dataset
# Replace "price" with your actual class label
attributes <- names(trainData1)[names(trainData1) != "price"]  # Exclude the class label
decision_tree <- construct_decision_tree(trainData1, "price", attributes)
```

```r
# Function to predict the class label for a single instance
predict_instance <- function(tree, instance) {
  if (length(tree$children) == 0) {
    # If the node is a leaf, return the predicted class label
    return(tree)
  }

  # Get the value of the instance for the attribute at the current node
  attribute_value <- instance[[tree$attribute]]

  # Traverse the tree based on the attribute value
  if (as.character(attribute_value) %in% names(tree$children)) {
    # Recursively predict for the corresponding subtree
    subtree <- tree$children[[as.character(attribute_value)]]
    return(predict_instance(subtree, instance))
  } else {
    # If the attribute value is not present in the training data, return the majority class
    return(names(sort(table(instance$class_label), decreasing = TRUE))[1])
  }
}

# Function to predict class labels for a test dataset
predict_dataset <- function(tree, test_data) {
  predictions <- sapply(1:nrow(test_data), function(i) {
    predict_instance(tree, test_data[i, ])
  })
  return(predictions)
}

# Example usage:
# Assuming testData is your test dataset
# Replace "price" with your actual class label
predicted_labels <- predict_dataset(decision_tree, testData1)
```
